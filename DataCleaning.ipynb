```python
import pandas as pd
import re #Regular Expression for the stopWords
```

```python
tweets = pd.read_csv("/content/sample_data/sentiment_tweets3.csv")
```

```python
nColumns = {'Index':'TweetID','message to examine':'RawText','label (depression result)':'Depression(0 or 1)'}
```

```python
tweets.rename(columns=nColumns, inplace=True)
```

```python
tweets.head()
```

```python
def remover_url(text):
    return re.sub(r'http\S+', '', text)

def remove_mentions(text):
    return re.sub(r'@\w+', '', text)

def remove_hashtags(text):
    return re.sub(r'#\w+', '', text)

def remove_special_characters(text):
    return re.sub(r'[^\w\s]', '', text)

def to_lowercase(text):
    return text.lower()
```

```python
tweets['RawText'] = tweets['RawText'].apply(remover_url)

tweets['RawText'] = tweets['RawText'].apply(remove_mentions)

tweets['RawText'] = tweets['RawText'].apply(remove_hashtags)

tweets['RawText'] = tweets['RawText'].apply(remove_special_characters)

tweets['RawText'] = tweets['RawText'].apply(to_lowercase)
```
